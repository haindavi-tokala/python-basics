```python
import tensorflow as tf

tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, input_shape=(1,), activation="relu", name="input_layer"),  # Changed units from 1 → 10
    tf.keras.layers.Dense(1, name="output_layer")  # Output layer remains unchanged
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

model.summary()  # View model architecture
```

```python
def plot_pred(train_data=xtrain,train_lables=ytrain,test_data=xtest,test_labels=ytest,predictions=ypred):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data,train_lables,c='b',label='training data')
  plt.scatter(test_data,test_labels,c='g',label='testing data')
  plt.scatter(test_data,predictions,c='r',label='predictions')
  plt.show()
```

```python
def mae(ytest,ypred):
   mae_metric = tf.keras.metrics.MeanAbsoluteError()
   mae_metric.update_state(ytest, tf.squeeze(ypred))  # Update state with true & predicted values
   mae_value = mae_metric.result().numpy()
   print("Mean Absolute Error:", mae_value)
   return mae_value
```

```python
mae(ytest,ypred)
```

```python
def mse(ytest,ypred):
   mse_metric = tf.keras.metrics.MeanSquaredError()
   mse_metric.update_state(ytest, tf.squeeze(ypred))  # Update state with true & predicted values
   mse_value = mse_metric.result().numpy()
   print("Mean squared Error:", mse_value)
   return mse_value
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

data=pd.read_csv('insurance.csv')
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
```

```python
ct=make_column_transformer((MinMaxScaler(),['age','bmi','children']),
                           (OneHotEncoder(handle_unknown='ignore'),['sex','smoker','region']))
x=data.drop('charges',axis=1)
y=data['charges']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
ct.fit(xtrain)
xtrain_norm=ct.transform(xtrain)
xtest_norm=ct.transform(xtest)
```

```python
xtrain_norm[0],xtest_norm[0]
```

```python
tf.random.set_seed(42)
mdl4=tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
mdl4.compile(loss=tf.keras.losses.mae,
             optimizer=tf.keras.optimizers.Adam(),
             metrics=['mae'])
mdl4.fit(xtrain_norm,ytrain,epochs=100)
```

```python
m4_eval=mdl4.evaluate(xtest_norm,ytest)
```
