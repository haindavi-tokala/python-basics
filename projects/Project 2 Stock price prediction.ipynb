```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
stk=input("enter code of the stock")
data=yf.download(stk,"2008-01-01","2021-01-18",auto_adjust=True)
data.head()
```

```python
from sklearn.model_selection import train_test_split
x=data.drop("Close",axis=1)
y=data["Close"]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
pred=lr.predict(xtest)
```

```python
pred
```

```python
ytest
```

```python
from sklearn.metrics import mean_squared_error,r2_score
def cal_metrics(ytest,pred):
    mse=mean_squared_error(ytest,pred)
    rmse=np.sqrt(mse)
    r2_scr=r2_score(ytest,pred)
    print(mse,rmse,r2_scr)  
cal_metrics(ytest,pred)

    
```

```python
from sklearn.linear_model import Lasso,Ridge
l=Lasso().fit(xtrain,ytrain)
r=Ridge().fit(xtrain,ytrain)
lpred=l.predict(xtest)
rpred=r.predict(xtest)
```

```python
cal_metrics(ytest,lpred)
```

```python
cal_metrics(ytest,rpred)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
svr=SVR()
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3)
grid.fit(xtrain,ytrain.values.ravel())
```

```python
svr=SVR(C=10,gamma=0.01,kernel='rbf')
svr.fit(xtrain,ytrain.values.ravel())
pred=svr.predict(xtest)
```
