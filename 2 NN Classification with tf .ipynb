```python
import numpy as np
def plot_dec_boundary(model,x,y):
  xmin,xmax=x[:,0].min()-0.1,x[:,0].max()+0.1
  ymin,ymax=x[:,1].min()-0.1,x[:,1].max()+0.1
  xx,yy=np.meshgrid(np.linspace(xmin,xmax,100),
                    np.linspace(ymin,ymax,100))
  xin=np.c_[xx.ravel(),yy.ravel()] #creates 2D stack
  ypred=model.predict(xin)
  if len(ypred[0])>1:
    print("multicls classification")
    ypred=np.argmax(ypred,axis=1).reshape(xx.shape)
  else:
    print("binary classification")
    ypred=np.round(ypred).reshape(xx.shape)
  plt.contourf(xx,yy,ypred,cmap=plt.cm.RdYlBu,alpha=0.7)
  plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())
```

```python
import itertools
from sklearn.metrics import confusion_matrix

def make_cm(ytest,ypred,clss=None,figsize=(10,10)):
  cm=confusion_matrix(ytest,ypred)
  cmnorm=cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
  ncls=cm.shape[0]

  fig,ax=plt.subplots(figsize=figsize)
  cax=ax.matshow(cm,cmap=plt.cm.Blues)

  if clss:
    labels=clss
  else:
    labels=np.arange(cm.shape[0])

  ax.set(title='Confusion Matrix',
        xlabel='predicted label',
        ylabel='True label',
        xticks=np.arange(ncls),
        yticks=np.arange(ncls),
        xticklabels=labels,
        yticklabels=labels)
  threshold=(cm.max()+cm.min())/2.
  for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
    plt.text(j,i,f'{cm[i,j]}({cmnorm[i,j]*100:.1f}%)',
            horizontalalignment='center',
            color='white' if cm[i,j]>threshold else "black",
            size=15)
```

1ï¸âƒ£ Input Layer: Flatten(input_shape=(28, 28))
Converts a 2D image (28x28 pixels) into a 1D vector of 784 values.
Example:

28x28 â†’ [x1, x2, x3, ..., x784]

2ï¸âƒ£ First Hidden Layer: Dense(128, activation='relu')
A fully connected (dense) layer with 128 neurons.
Uses ReLU (Rectified Linear Unit) activation:

ReLU(x) = max(0, x)
Helps the network learn complex patterns.

3ï¸âƒ£ Batch Normalization: BatchNormalization()
Normalizes activations across a batch.
Normalizes activations per mini-batch.
Stabilizes training and speeds up convergence.
Helps prevent the vanishing gradient problem.

4ï¸âƒ£ Dropout: Dropout(0.2)
Drops 20% of neurons randomly during training.
Prevents overfitting by forcing the network to generalize.

5ï¸âƒ£ Second Hidden Layer: Dense(64, activation='relu')
Similar to the first hidden layer but with 64 neurons.
Helps extract deeper patterns.

6ï¸âƒ£ Batch Normalization & 7ï¸âƒ£ Dropout (Again)
Batch Normalization ensures stable training.
Dropout (20%) prevents overfitting.

8ï¸âƒ£ Output Layer: Dense(10, activation='softmax')
10 neurons (one for each class if class labels are 0-9).
Uses Softmax activation, which converts outputs into probabilities:

softmax(xi) = exp(xi) / Î£ exp(xj)
Ensures the sum of probabilities across all classes = 1.

âœ¨ Summary


Flatten
Converts 2D input (28x28) into a 1D vector (784 values).
Dense(128, ReLU)

First hidden layer (128 neurons). Learns patterns.

BatchNormalization	Stabilizes activations, speeds up training.

Dropout(0.2)	Drops 20% of neurons to prevent overfitting.
Dense(64, ReLU)	Second hidden layer (64 neurons). Extracts deeper features.

BatchNormalization	Again normalizes for stable training.

Dropout(0.2)	Again prevents overfitting.

Dense(10, Softmax)	Final output layer (10 classes). Converts to probabilities.
