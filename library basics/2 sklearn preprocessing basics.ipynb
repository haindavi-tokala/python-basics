```python
data1=pd.read_csv('car-sales-extended-missing-data.csv')
data1.head()
```

```python
data1.isna().sum()
```

```python
data1.dropna(subset=['Price'],inplace=True)
data1.isna().sum()
```

```python
x=data1.drop('Price',axis=1)
y=data1['Price']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
cat_imp=SimpleImputer(strategy='constant',fill_value='missing')
door_imp=SimpleImputer(strategy='constant',fill_value=4)
num_imp=SimpleImputer(strategy='mean')

cat=['Make','Colour']
door=['Doors']
num=['Odometer (KM)']

imp=ColumnTransformer([('cat_imp',cat_imp,cat),('door_imp',door_imp,door),('num_imp',num_imp,num)])

filledx=imp.fit_transform(x)
filledx
```

```python
upddata=pd.DataFrame(filledx,columns=['Make','Colour','Odometer (KM)','Doors'])
upddata.head()
```

```python
upddata.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
cols=['Make','Colour','Doors']
ohe=OneHotEncoder()
ct=ColumnTransformer([('ohe',ohe,cols)],remainder='passthrough')
xtrans=ct.fit_transform(upddata)
xtrans
```

```python
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xtrans,y,test_size=0.2)
clf=RandomForestRegressor(n_estimators=100)
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
```
