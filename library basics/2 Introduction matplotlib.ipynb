```python
import pandas as pd
dis=pd.read_csv("heart-disease.csv")
dis.head()
```

```python
dis['age'].plot.hist(bins=10),dis['age'].plot.hist(bins=30)
```

```python
dis.plot.hist(figsize=(10,30),subplots=True)
```

```python
over50=dis[dis['age']>50]
over50.head()
```

```python
over50.plot(kind='scatter',x='age',y='chol',c='target')
```

```python
fig,ax=plt.subplots(figsize=(8,4))
over50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
ax.set_xlim([45,100])
```

```python
fig,ax=plt.subplots(figsize=(8,4))
sc=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'])
ax.set(title='heart dis and chol level',
       xlabel='age',
       ylabel='chol')
ax.legend(*sc.legend_elements(),title='Target')
ax.axhline(over50['chol'].mean(),linestyle='--')
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

sc=ax1.scatter(x=over50['age'],y=over50['chol'],c=over50['target'])
ax1.set(title='heart dis and chol level',
       xlabel='age',
       ylabel='chol')
ax1.legend(*sc.legend_elements(),title='Target')
ax1.axhline(over50['chol'].mean(),linestyle='--')

sc=ax2.scatter(x=over50['age'],y=over50['thalach'],c=over50['target'])
ax2.set(title='heart dis and chol level',
       xlabel='age',
       ylabel='talach')
ax2.legend(*sc.legend_elements(),title='Target')
ax2.axhline(over50['thalach'].mean(),linestyle='--')

fig.suptitle("heart disease analysis",fontsize=16,fontweight='bold')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig,ax=plt.subplots(figsize=(8,4))
sc=ax.scatter(x=over50['age'],y=over50['chol'],c=over50['target'],cmap='plasma')
ax.set(title='heart dis and chol level',
       xlabel='age',
       ylabel='chol')
ax.legend(*sc.legend_elements(),title='Target')
ax.axhline(over50['chol'].mean(),linestyle='--')
```
