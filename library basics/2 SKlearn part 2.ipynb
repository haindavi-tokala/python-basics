```python
def eval_pred(ytest,ypred):
    acc=accuracy_score(ytest,ypred)
    prec=precision_score(ytest,ypred)
    recall=recall_score(ytest,ypred)
    f1=f1_score(ytest,ypred)
    metric_dict={'accuracy':round(acc,2),
                 'precision':round(prec,2),
                 'recall':round(recall,2),
                 'f1':round(f1,2)}
    print(f'acc={acc*100:.2f}%')
    print(f'prec={prec*100:.2f}')
    print(f'recall={recall*100:.2f}')
    print(f'f1={f1*100:.2f}')
    return metric_dict
          
```

```python
from sklearn.ensemble import RandomForestClassifier
np.random.seed(42)
data=pd.read_csv('heart-disease.csv')
data.head()
```

```python
datas=data.sample(frac=1)
```

```python
x=datas.drop('target',axis=1)
y=datas['target']
```

```python
train_split=round(0.7*len(datas))
```

```python
valid_split=round(train_split+0.15*len(datas))
```

```python
xtrain,ytrain=x[:train_split],y[:train_split]
```

```python
xvalid,yvalid=x[train_split:valid_split],y[train_split:valid_split]
```

```python
xtest,ytest=x[valid_split:],y[valid_split:]
```

```python
len(xtrain),len(xvalid),len(xtest)
```

```python
clf=RandomForestClassifier(n_estimators=10)
```

```python
clf.fit(xtrain,ytrain)
```

```python
ypred=clf.predict(xvalid)
```

```python
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score
baseline_metric=eval_pred(yvalid,ypred)
```

```python
np.random.seed(42)
clf2=RandomForestClassifier(n_estimators=100)
clf2.fit(xtrain,ytrain)
ypred2=clf2.predict(xvalid)
clf2metrics=eval_pred(yvalid,ypred2)
```

```python
np.random.seed(42)
clf3=RandomForestClassifier(n_estimators=100,max_depth=100)
clf3.fit(xtrain,ytrain)
ypred3=clf3.predict(xvalid)
baseline_metric=eval_pred(yvalid,ypred3)
```
